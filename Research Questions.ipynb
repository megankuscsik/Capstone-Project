2. [Does the weather affect stock market volatility?](https://www.researchgate.net/publication/222786068_Does_the_weather_affect_stock_market_volatility) <br/>
Symeonidis, Lazaros, et al. “Does the Weather Affect Stock Market Volatility?” Finance Research Letters, vol. 7, no. 4, 2010, pp. 214–223., doi:10.1016/j.frl.2010.05.004. <br/>
    - paper investigates the empirical association between **stock market volatility** and investor mood-proxies related to the weather (cloudiness, temperature and precipitation) and the environment (nighttime length) 
    - results suggest that cloudiness and length of nighttime are inversely related to historical, implied and realized measures of volatility 
    - strength of association seems to vary with the location of an exchange on Earth with respect to the equator
    - Weather deviations from seasonal norms and dummies representing extreme weather conditions do not offer additional explanatory power in our datasets
    
    
3. [The Impact of Climate Change Information: New Evidence from the Stock Market](https://static1.squarespace.com/static/55e8ab64e4b0b55649c4ab64/t/55edee9ee4b0d0e40f4e3a83/1441656478212/Beatty_Shimshack_2010.pdf) <br/>
Beatty, Timothy, and Jay P Shimshack. “The Impact of Climate Change Information: New Evidence from the Stock Market.” The B.E. Journal of Economic Analysis &amp; Policy, vol. 10, no. 1, 2010, doi:10.2202/1935-1682.2374. <br/>
    - paper’s key contribution is providing early evidence on the relationship between financial performance and climate-related environmental behavior
    - our study examines climate change information and a plausibly exogenous event releasing company climate ratings
    - find that the release of these climate ratings had statistically significant and large impacts on stock market returns
    - We find that the event information had an economically important and statistically significant impact on capital market returns
    - Poorly rated firms suffered market penalties, but we find limited benefits for firms receiving good ratings 
    - We also uncover suggestive evidence that the economic mechanism driving our results is not a direct consumer demand effect
    - evaluated **stock price at close and daily returns**
    
    
4. [Climate Change and Price Volatility: Can We Count on the ASEAN Plus Three Emergency Rice Reserve?](https://www.adb.org/sites/default/files/publication/29971/adb-wp-24-climate-change-price-volatility.pdf)  <br/>
Briones, Roehlano  M., et al. “Climate Change and Price Volatility: Can We Count on the ASEAN Plus Three Emergency Rice Reserve?” ADB Sustainable Development Working Paper Series , no. 24, Aug. 2012, doi:WPS124955-3. <br/>
    - found that natural calamities (changes to normal weather) have an impact on the growth of rice
    - less product means less product to sell, therefore driving up the price (creating **volatility**)
    - have rice reserves to help feed demands when poor growing seasons occur (try and control the price)
    - wihtout the rice reserves, price would increase due to rice available not being able to meet demand
   

5. [Climate Risks and Market Efficiency](https://www.q-group.org/wp-content/uploads/2017/02/Climate-Risks-and-Market-Efficiency.pdf) <br/>
Hong, Harrison, et al. “Climate Risks and Market Efficiency.” Nov. 2016, doi:10.3386/w22890. <br/>
    - study investigates whether stock market price risks (volatility) brought on or exacerbated by climate change
    - focus on drought, the most damaging natural disaster for crops and food-company cash flows
    - we show that prolonged drought spells in a country, as measured by a 3-year moving average of PDSI, forecast poor returns for a portfolio comprised of food stocks in that country (**creates volatility**)
    - predictability is stronger in countries with previously temperate climates and little history of droughts
